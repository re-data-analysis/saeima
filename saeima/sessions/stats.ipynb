```python
import pandas as pd

from saeima import PROJECT_DIR

VOTES_FILE_PATH = PROJECT_DIR / "data/sessions/votes.csv"
```

```python
votes_df = pd.read_csv(VOTES_FILE_PATH)
```

```python
len(votes_df)
```

```python
(votes_df.result.value_counts())
```

```python
pd.set_option("max_colwidth", None)

exclude_strings = ["iekļaušanu Saeimas sēdes darba kārtībā", "Par priekšlikumu Nr"]

# exclude rows with vote_motive containing any of the strings
_votes_df = votes_df[~votes_df.vote_motive.str.contains("|".join(exclude_strings))]

# check how many are left
len(_votes_df)
```

```python
_votes_df.columns
```

```python
(
    _votes_df.query("result == 'Atturas'")
    .groupby(["name", "surname"])
    .agg(counts=("result", "count"))
    # remove index
    .reset_index()
    .sort_values("counts", ascending=False)
)
```
